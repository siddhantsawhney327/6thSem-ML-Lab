```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/hiring.csv")

# Rename columns
df.rename(columns={
    'test_score(out of 10)': 'test_score',
    'interview_score(out of 10)': 'interview_score',
    'salary($)': 'salary'
}, inplace=True)

# Convert word numbers to digits
experience_map = {
    'zero':0,
    'one':1,
    'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    'seven':7,
    'eight':8,
    'nine':9,
    'ten':10,
    'eleven':11
}

df['experience'] = df['experience'].map(experience_map)

# Fill missing values
df['experience'] = df['experience'].fillna(0)
df['test_score'] = df['test_score'].fillna(df['test_score'].mean())
df['interview_score'] = df['interview_score'].fillna(df['interview_score'].mean())

print("\nProcessed Data:")
print(df)

# Features and target
X = df[['experience', 'test_score', 'interview_score']]
y = df['salary']

# Train model
model = LinearRegression()
model.fit(X, y)

# Predictions
salary1 = model.predict([[2, 9, 6]])
salary2 = model.predict([[12, 10, 10]])

print("\nPredicted Salaries:")
print("2 yrs, 9 test, 6 interview:", salary1[0])
print("12 yrs, 10 test, 10 interview:", salary2[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/1000_Companies.csv")

print(df.head())

# One-hot encode State column
df = pd.get_dummies(df, columns=['State'], drop_first=True)

# Features
X = df.drop('Profit', axis=1)

# Target
y = df['Profit']

# Create model
model = LinearRegression()

# Train
model.fit(X, y)

# Florida becomes: State_New York=0 , State_California=0
prediction = model.predict([[91694.48, 515841.3, 11931.24, 0, 0]])

print("\nPredicted Profit:")
print(prediction[0])
```
