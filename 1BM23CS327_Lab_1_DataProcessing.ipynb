```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Diabetes Dataset
diabetes_df = pd.read_csv("diabetes.csv")

# Adult Income Dataset
adult_df = pd.read_csv("adult.csv")

print(diabetes_df.head())
print(adult_df.head())
```

```python
# Columns where 0 is invalid
cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# Replace 0 with NaN
diabetes_df[cols_with_zero] = diabetes_df[cols_with_zero].replace(0, np.nan)

# Fill missing values with column mean
diabetes_df.fillna(diabetes_df.mean(), inplace=True)

print(diabetes_df.isnull().sum())
```

```python
print(diabetes_df.columns)
```

```python
import numpy as np

# Columns where 0 indicates missing values
cols_with_zero = [
    'Urea', 'Cr', 'HbA1c', 'Chol', 'TG',
    'HDL', 'LDL', 'VLDL', 'BMI'
]

# Replace 0 with NaN
diabetes_df[cols_with_zero] = diabetes_df[cols_with_zero].replace(0, np.nan)

# Fill missing values with mean
diabetes_df[cols_with_zero] = diabetes_df[cols_with_zero].fillna(
    diabetes_df[cols_with_zero].mean()
)

# Verify
print(diabetes_df.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

diabetes_df['Gender'] = le.fit_transform(diabetes_df['Gender'])
diabetes_df['CLASS'] = le.fit_transform(diabetes_df['CLASS'])

print(diabetes_df.head())
```

```python
numeric_cols = diabetes_df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    Q1 = diabetes_df[col].quantile(0.25)
    Q3 = diabetes_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    diabetes_df[col] = np.where(
        diabetes_df[col] > upper, upper,
        np.where(diabetes_df[col] < lower, lower, diabetes_df[col])
    )
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()
diabetes_minmax = pd.DataFrame(
    minmax.fit_transform(diabetes_df),
    columns=diabetes_df.columns
)
```

```python
from sklearn.preprocessing import StandardScaler

standard = StandardScaler()
diabetes_standard = pd.DataFrame(
    standard.fit_transform(diabetes_df),
    columns=diabetes_df.columns
)
```
