```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/canada_per_capita_income.csv")

print(df.head())

# Features and target
X = df[['year']]        # Independent variable
y = df['per capita income (US$)']   # Dependent variable

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict for year 2020
prediction_2020 = model.predict([[2020]])

print("\nPredicted per capita income in 2020:")
print(prediction_2020[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/salary.csv")

# Show original data
print("Original Data:")
print(df)

# Remove rows with missing values
df = df.dropna()

print("\nAfter removing NaN values:")
print(df)

# Independent & dependent variables
X = df[['YearsExperience']]
y = df['Salary']

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict salary for 12 years experience
predicted_salary = model.predict([[12]])

print("\nPredicted Salary for 12 years experience:")
print(predicted_salary[0])

```
